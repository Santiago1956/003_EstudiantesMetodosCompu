```python
def choque_pi(m1, m2, v1, v2):

    choq = 0

    if v1 < v2:

        return choq

    while True:

        v1_n = ((m1 - m2) * v1 + 2 * m2 * v2) / (m1 + m2)
        v2_n = ((m2 - m1) * v2 + 2 * m1 * v1) / (m1 + m2)

        v1 = v1_n
        v2 = v2_n
        choq += 1


        if v1 < 0:

            v1 = -v1
            choq += 1

        if v1 <= v2:

            break

    return choq
```

```python
choque_pi(1,100,0,-1)
```
