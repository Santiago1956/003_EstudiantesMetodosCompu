3. **Diagonalización**.

Se sabe que un sistema físico es gobernado por las ecuaciones,



$$
\begin{matrix}
x' &=& 2x&-y&+z, \\
y' &=&   &3y&-1z, \\
z' &=& 2x&+y&+3z,
\end{matrix}
$$

donde $x',y',z'$ son las derivadas, en forma vectorial esto se puede representar como $\mathbf{r}'=A\mathbf{r}$ que tiene solución analítica general, $\mathbf{r} = e^{At} \mathbf{r}(0)$.<br>
a) Diagonalize $A$.<br>



```python
import sympy as sp
import numpy as np
import scipy as scp
```

```python
A = np.array([
     [2,-1,1],
     [0,3,-1],
     [2,1,3]])
```

```python
#Valores propios
eigenvalor, eigenvector = scp.linalg.eig(A)

#Diagonal
diagonal = np.diag(eigenvalor)

#Matriz P
p = eigenvector
p_inv = np.linalg.inv(p)

#Construimos
A_n = p@diagonal@p_inv
```

```python
print(f'La matriz diagonal de A es:\n{diagonal}\nContiene valores imaginarios')
```

```python
print(f'La matriz P de A es:\n{p}\nContiene valores imaginarios')
```

```python
print(f'La matriz diagonal de P inversa de A es:\n{p_inv}\nContiene valores imaginarios')
```
