```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def g(y,k):
  return -k*y
```

```python
a = 0
b = 10
h = 0.01
k = 0.5
y0 = 0.8
```

```python
N = int((b - a) / h)
t = np.linspace(a, b, N)
y = np.zeros(N)
y[0] = y0

#Método de Euler
for i in range(N - 1):
    y[i+1] = y[i] + h * g(y[i], k)
```

```python
# Graficar la solución
plt.plot(t, y, label="f(t)")
plt.xlabel("Tiempo(t)")
plt.ylabel("Decae(t)")
plt.title("Decaimiento")
plt.legend()
plt.grid()
plt.show()
```
